```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import sqlite3

conn = sqlite3.connect('jsondata.db')
df = pd.read_sql_query("SELECT * FROM USINDUSTRY", conn)
conn.close()
```

```python
df2 = df.replace('', np.nan)
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2[df2["end_year"] == df2["end_year"].max()]
```

```python
df2["end_year"].replace(to_replace = np.NAN,value = "Undefined",inplace=True)
```

```python
df2["start_year"].replace(to_replace=np.NAN,value = "Undefined",inplace=True)
```

```python
df2["intensity"].replace(np.NAN,0,inplace=True)
```

```python
df2.info()
```

```python
df2.columns.tolist()
```

```python
columns = df2.dtypes.to_dict()
```

```python
columns
```

```python
for column in columns:
    if columns[column] == 'object':
        df2[column].replace(to_replace=np.NAN,value = "Undefined",inplace=True)
    else:
        df2[column].replace(to_replace=np.NAN,value = 0,inplace=True)
```

## Graphs that will go into the website

```python
filter = input('Enter a filter: ')
df2[filter].value_counts().plot(kind='bar')
data = df2[filter].value_counts()
print(type(data))
print(data.to_json())
```

```python
filter = input("Enter filter: ")
constraint = input("Enter constraint: ")
filter2 = input("show chart based on :")
df2[df2[filter] == constraint][filter2].value_counts().plot(kind='bar')
data = df2[df2[filter] == constraint][filter2].value_counts()
print(data.to_json())
```

```python
filter1 = input("Enter the first filter: ")
filter2 = input("Enter the second filter: ")
df2.groupby([filter1,filter2]).size().unstack().plot(kind='bar',stacked=True)
data = df2.groupby([filter1,filter2]).size().unstack()
print(data.to_json())
```
